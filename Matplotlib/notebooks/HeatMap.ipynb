```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
from dotenv import load_dotenv
load_dotenv(override=True)
plt.style.use("fivethirtyeight")
```

## HeatMap

```python
ipl = pd.read_csv(os.getenv("IPL_BALL_FILE_PATH"))
ipl.head()
```

```python
ipl_df = ipl[(ipl["ballnumber"].isin([1,2,3,4,5,6])) & (ipl["batsman_run"]==6)]
ipl_grid = ipl_df.pivot_table(index="overs",columns="ballnumber",values="batsman_run",aggfunc='count')

plt.figure(figsize=(6,8))
plt.imshow(ipl_grid, cmap='viridis')
plt.xticks(np.arange(0,6), np.arange(1,7))
plt.yticks(np.arange(0,20),np.arange(1,21))
plt.colorbar()
```
