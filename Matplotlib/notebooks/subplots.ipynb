```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
from dotenv import load_dotenv
load_dotenv(override=True)
plt.style.use("ggplot")
```

## Subplots

```python
iris = sns.load_dataset("iris")
iris.sample(5)
```

```python
iris["species"] = iris["species"].replace({"setosa":0,"versicolor":1,"virginica":2})

fig,ax = plt.subplots()
ax.scatter(iris["sepal_width"], iris["petal_width"], c=iris["species"],marker="+")
ax.set_title("Iris: sepal width vs petal width")
ax.set_xlabel("sepal width")
ax.set_ylabel("petal width")

# fig.show()
```

```python
# multiple graphs
batters = pd.read_csv(os.getenv("BATTER_FILE_PATH"))
batters.sample(5)
```

```python
fig,ax = plt.subplots(2,1,figsize=(8, 10), sharex=True )  
ax[0].scatter(batters["avg"], batters["strike_rate"])
ax[0].set_title("avg vs strike rate ")

ax[1].scatter(batters["avg"], batters["runs"],color="#0A9F21")
ax[1].set_title("avg vs runs")
```

```python
fig, ax = plt.subplots(2,2, figsize=(8,8)) # here ax is 2D array

ax[0,0].scatter(batters["avg"], batters["strike_rate"])
ax[0,0].set_title("avg vs strike rate ")

ax[0,1].scatter(batters["avg"], batters["runs"])
ax[0,1].set_title("avg vs runs ")

ax[1,0].hist(batters["avg"])
ax[1,0].set_title("avg ")

ax[1,1].hist(batters["strike_rate"])
ax[1,1].set_title("strike_rate ")
```
