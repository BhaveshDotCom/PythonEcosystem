```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
from dotenv import load_dotenv
load_dotenv(override=True)
plt.style.use("ggplot")
```

## Scatter Plot

```python
x = np.linspace(-10,10,50)
y = 10*x+3+np.random.randint(3,300,50)

plt.scatter(x,y)
```

```python
batter = pd.read_csv(os.getenv("BATTER_FILE_PATH"))
batter= batter.head(50)
plt.scatter(batter["avg"], batter["strike_rate"], marker="o")
plt.title("Average vs Strike Rate Top 50") 
plt.xlabel("Average")
plt.xlabel("Strike Rate")
```
