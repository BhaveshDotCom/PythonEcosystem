```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
from dotenv import load_dotenv
load_dotenv(override=True)
plt.style.use("ggplot")
```

## Bar Chart

```python
sns.get_dataset_names()
```

```python
flights = sns.load_dataset("flights").groupby("month").agg({"passengers":"sum"}).reset_index()
flights
```

```python
plt.bar(flights["month"], flights["passengers"],color="#1d63b3")
```

```python
batsmans = pd.read_csv(os.getenv("TOP_5_BATSMAN_FILE_PATH"))
batsmans
```

##### Multiple Bar Chart

```python
plt.bar(np.arange(batsmans.shape[0])-0.2, batsmans["2015"], width=0.2)
plt.bar(np.arange(batsmans.shape[0]), batsmans["2016"], width=0.2)
plt.bar(np.arange(batsmans.shape[0])+0.2, batsmans["2017"], width=0.2)

plt.xticks(np.arange(batsmans.shape[0]), batsmans["batsman"])
plt.xticks(rotation=50)

# Save 
plt.savefig("IPL")
```

```python
plt.bar(batsmans["batsman"], batsmans["2015"], label="2015")
plt.bar(batsmans["batsman"], batsmans["2016"], bottom=batsmans["2015"], label="2016")
plt.bar(batsmans["batsman"], batsmans["2017"], bottom=batsmans["2015"]+batsmans["2016"], label="2017")

plt.legend()

plt.show()
```
