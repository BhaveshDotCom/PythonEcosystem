##### 2D Line Plot
A 2D Line Plot is a graph that displays the relationship or trend between two variables by connecting data points with straight lines on a two-dimensional coordinate system (X-axis and Y-axis).

A **2D Line Plot** is used to:

- **Show trends over time**
  - Stock prices over days
  - Monthly sales
  - Daily temperature
  - Website visitors per day

- **Compare continuous data**
  - Distance vs. Time
  - Speed vs. Time
  - Height vs. Age

- **Visualize changes**
  - Population growth
  - Company revenue over the years
  - Rainfall over months

- **Identify patterns**
  - Increasing trend
  - Decreasing trend
  - Seasonal variations
  - Peaks and dips in data

```python
roko_runs = pd.read_csv(os.getenv("RUNS_ROKO_FILE_PATH"))
roko_runs
```

```python
plt.plot(roko_runs["index"], roko_runs[["V Kohli", "RG Sharma"]])
```

```python
plt.plot(roko_runs["index"], roko_runs["RG Sharma"],color="#4D900E", linewidth=4)
plt.plot(roko_runs["index"], roko_runs["V Kohli"], color="#0E4D90", linestyle='dashdot')
plt.title("Rohit vs Kohli")
plt.xlabel("IPL Seasom")
plt.ylabel("Runs")
```

```python
plt.plot(roko_runs["index"], roko_runs["RG Sharma"], linewidth="4", marker='d', markersize=10 )
plt.plot(roko_runs["index"], roko_runs["V Kohli"], color="#0E4D90", marker='o')
plt.title("Rohit vs Kohli")
plt.xlabel("IPL Seasom")
plt.ylabel("Runs")
```

```python
plt.plot(roko_runs["index"], roko_runs["RG Sharma"], label="Rohit Sharma", marker="o")
plt.plot(roko_runs["index"], roko_runs["V Kohli"], label="Virat Kohli")
plt.title("Rohit vs Kohli")
plt.legend()
plt.xlabel("IPL Seasom")
plt.ylabel("Runs")
plt.show()
```
