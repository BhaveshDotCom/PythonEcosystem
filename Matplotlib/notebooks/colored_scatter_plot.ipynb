```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
from dotenv import load_dotenv
load_dotenv(override=True)
plt.style.use("ggplot")
```

```python
iris = sns.load_dataset("iris")
iris["species"] = iris["species"].replace({"setosa":0, "versicolor":1, "virginica":2})
iris.sample(5)
```

```python
# plt.figure(figsize=(5,9))
plt.scatter(iris["sepal_length"], iris["petal_length"],c=iris["species"])
plt.xlabel("sepal_length")
plt.ylabel("petal_length")
plt.colorbar()
```

```python
# Annotations
batters = pd.read_csv(os.getenv("BATTER_FILE_PATH"))
batter = batters.head(100).sample(25,random_state=42)
batter
```

```python
plt.figure(figsize=(18,10))
plt.scatter(batter["avg"], batter["strike_rate"], s=batter["runs"])

# label
for i in range(batter.shape[0]):
    plt.text(batter["avg"].values[i],batter["strike_rate"].values[i],batter["batter"].values[i],fontsize=20)
```

```python
plt.figure(figsize=(18,10))
plt.scatter(batter["avg"], batter["strike_rate"], s=batter["runs"])

plt.axhline(130)
plt.axvline(30,color="green")
# label
for i in range(batter.shape[0]):
    plt.text(batter["avg"].values[i],batter["strike_rate"].values[i],batter["batter"].values[i],fontsize=20)
```
