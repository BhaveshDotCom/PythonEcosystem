```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
from dotenv import load_dotenv
load_dotenv(override=True)
plt.style.use("fivethirtyeight")
```

## 3D Plots

```python
batters = pd.read_csv(os.getenv("BATTER_FILE_PATH"))
batters.sample(5)
```

##### 3D scatter

```python
fig = plt.figure(figsize=(7,5))

ax = plt.subplot(projection="3d")
ax.scatter3D(batters["avg"],batters["runs"], batters["strike_rate"])
ax.set_title("Batters Analysis")
ax.set_xlabel("avg")
ax.set_ylabel("runs")
ax.set_zlabel("strike rate")
```

##### 3D Line

```python
np.random.seed(42)
batter_df = batters.sample(4)

fig = plt.figure(figsize=(7,5))

ax = plt.subplot(projection="3d")
ax.plot3D(batter_df["avg"],batter_df["runs"], batter_df["strike_rate"],color="brown")
ax.set_title("batters Analysis")
ax.set_xlabel("avg")
ax.set_ylabel("runs")
ax.set_zlabel("strike rate")
```

##### 3D Surface

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
xx,yy = np.meshgrid(x,y)
z = xx**2 + yy**2 
plt.figure(figsize=(7,5))
ax = plt.subplot(projection="3d")
graph = ax.plot_surface(xx,yy,z,cmap="viridis")
fig.colorbar(graph)
```

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
xx,yy = np.meshgrid(x,y)
z = np.cos(xx)**2 + np.sin(yy)**2
plt.figure(figsize=(7,5))
ax = plt.subplot(projection="3d")
graph = ax.plot_surface(xx,yy,z,cmap="viridis")
fig.colorbar(graph)
```

##### Contour Plot

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
xx,yy = np.meshgrid(x,y)
z = xx**2 + yy**2 
plt.figure(figsize=(7,5))
ax = plt.subplot()
graph = ax.contourf(xx,yy,z,cmap="viridis")
fig.colorbar(graph)
```
