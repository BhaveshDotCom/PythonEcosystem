```python
import seaborn as sns
import matplotlib.pyplot as plt
```

## Categorical Plot

```python
# Datasets
tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")
```

### Categorical Scatter Plot

```python
# stripplot
# axex level function
sns.stripplot(data=tips, x='day', y='total_bill', hue='day', jitter=False)
```

```python

sns.catplot(kind="strip", data=tips, x="day", y="total_bill", hue="sex")
```

```python
sns.catplot(kind="swarm",data=tips, x="day", y="total_bill", hue="sex")
```

### Categorical Distribution Plot

```python
# sns.catplot(kind="box",data=tips, x="day", y="total_bill", hue="sex")
ax = sns.boxplot(data=tips, x="day", y="total_bill", hue="sex")
```

```python
sns.boxplot(data=tips, y="tip")
```

```python
sns.violinplot(data=tips,x="sex", y="total_bill", hue="day")
# sns.violinplot(data=tips,x="sex", y="total_bill", hue="day", split=True)
```

```python
sns.violinplot(data=tips,x="sex", y="total_bill")
sns.swarmplot(data=tips,x="sex", y="total_bill")
```
