```python
import numpy as np
import pandas as pd
import seaborn as sns
import os
import plotly.express as px
from dotenv import load_dotenv
load_dotenv(override=True)
```

```python
tips = sns.load_dataset("tips")
```

## histplot

```python
# hisplot (axes level)
sns.histplot(data=tips, x="total_bill",bins=30)
```

```python
# hisplot (figure level)
sns.displot(kind="hist", data=tips, x="total_bill", hue="sex", element="step")
```

```python
# hist on categorical data
sns.displot(kind="hist", data=tips, x="day") # count plot
```

```python
titanic = sns.load_dataset("titanic")
titanic
```

```python
sns.displot(kind="hist",data=titanic, x="age", hue="survived")
```

```python
# Bivariate
sns.displot(data=tips,kind="hist", x="total_bill", y="tip")
```

## kdeplot
A KDE plot (Kernel Density Estimate plot) is a visualization tool that shows the distribution of data using a smooth, continuous line, rather than rigid bars like a histogram

It estimates the probability density function of a continuous variable, making it easier to see the overall shape, peaks, and spreads of your data.

```python
# sns.kdeplot(data=tips, x="total_bill")
sns.displot(kind="kde",data=tips, x="total_bill", hue="sex", fill=True)
```

```python
# Bivariate
sns.kdeplot(data=tips, y="total_bill", x="tip")
```

## rugplot
visualization that draws small vertical ticks along an axis

```python
sns.displot(data=tips, x="total_bill",kind="kde", fill=True, rug=True)
```

```python
sns.displot(data=tips, x="total_bill",kind="kde", fill=True, rug=True, y="tip")
```
