```python
import numpy as np
import pandas as pd
import seaborn as sns
import os
import plotly.express as px
from dotenv import load_dotenv
load_dotenv(override=True)
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
# scatter plot (axes level function)
sns.scatterplot(data=tips, x="total_bill", y="tip")
```

```python
# scatter plot (figure level function)
sns.relplot(data=tips, x="total_bill", y="tip", kind="scatter", hue="sex", style="time", size="size")
```

```python
gap = px.data.gapminder()
gap.sample(7)
india_data = gap[gap["country"] == "India"]
```

```python
# line plot (axes level)
sns.lineplot(data=india_data, x="year", y="lifeExp")
# line plot (figure level)
sns.relplot(data=india_data, x="year", y="lifeExp", kind="line", )
```

```python
country_df = gap[gap["country"].isin(["India", "China", "Japan"])]
sns.lineplot(data=country_df, x="year",y="lifeExp", hue="country")
```

```python
country_df = gap[gap["country"].isin(["India", "China", "Japan", "Germany", "Brazil"])]
sns.relplot(data=country_df, x="year",y="lifeExp",kind="line", hue="country", style="continent", size="continent")
```

```python
sns.relplot(data=tips, x="total_bill", y="tip", col="sex",row="smoker", hue="sex",height=3,aspect=1.2)
sns.relplot(data=tips, x="total_bill", y="tip", col="sex",col_wrap=1, hue="sex",height=3,aspect=1.2)
```
