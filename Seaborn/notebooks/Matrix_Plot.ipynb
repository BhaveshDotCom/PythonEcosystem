```python
import numpy as np
import pandas as pd
import seaborn as sns
import os
import plotly.express as px
from dotenv import load_dotenv
import matplotlib.pyplot as plt
import scipy
load_dotenv(override=True)
```

```python
countries_data = px.data.gapminder()
countries_data
```

## Heatmap

```python
countries_pivot = countries_data.pivot(index="country", columns="year", values="lifeExp")
```

```python
# axes level
plt.figure(figsize=(6,6))
sns.heatmap(data=countries_pivot)

plt.tight_layout()
plt.savefig("Countries", bbox_inches="tight")
```

```python
europe_data = countries_data[countries_data["continent"] == "Europe"].pivot(index="country", columns="year", values="lifeExp")
sns.heatmap(data=europe_data, annot=True, linewidths=0.5, cmap="summer", annot_kws={"size": 7})
```

## clustermap
a specialized Seaborn matrix plot that automatically performs hierarchical clustering on your dataset's rows and columns. It reorders the data to group similar patterns together and visualizes them as a color-coded heatmap bordered by tree-like diagrams called dendrograms.

```python
iris = sns.load_dataset("iris")
iris
```

```python
sns.clustermap(iris.iloc[:,[0,1,2,3]])
```
