```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")
```

```python
sns.catplot(kind="violin" ,data=tips, x="sex", y="total_bill", col="day", row="sex", hue="sex")
```

```python
facet_grid_object = sns.FacetGrid(data=tips, col="day", row="sex", hue="day")
facet_grid_object.map(sns.violinplot, "sex", "total_bill")
facet_grid_object.add_legend()
```

```python
# Pair Plot
iris = sns.load_dataset("iris")
sns.pairplot(data=iris, hue="species")
```

```python
# pairgrid
g = sns.PairGrid(data=iris, hue="species")
g.map(sns.scatterplot)
```
