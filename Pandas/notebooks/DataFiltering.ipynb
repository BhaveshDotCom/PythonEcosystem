## Data Filtering

```python
import numpy as np
import pandas as pd
import os
import dotenv
dotenv.load_dotenv()
```

```python
ipl_data = pd.read_csv(os.getenv("IPL_DATA_FILE_PATH"))
ipl_data
```

```python
# Find all ipl winner
ipl_data[ipl_data["MatchNumber"] == "Final"][["Season","WinningTeam"]]
```

```python
# Number of superover
ipl_data[ipl_data["SuperOver"] == "Y"].shape[0]
```

```python
# how many matches has CSK won in kolkata
csk_kolkata = ipl_data[(ipl_data["City"] == "Kolkata") & (ipl_data["WinningTeam"] == "Chennai Super Kings")]
csk_kolkata.shape[0]
```

```python
# toss win, match win ?
(ipl_data[ipl_data["TossWinner"] == ipl_data["WinningTeam"]].shape[0] / ipl_data.shape[0])*100
```

```python
movies = pd.read_csv(os.getenv("MOVIES_FILE_PATH"))
movies
```

```python
# movies with rating >8 and votes > 10000
movies[(movies["imdb_rating"] > 8) & (movies["imdb_votes"] > 1000)].shape[0]
```

```python
# action movies with rating >7.5
movies[(movies["genres"].str.split('|').apply(lambda genre : "Action" in genre)) & (movies["imdb_rating"] > 7.5)].shape[0]
# movies[(movies["genres"].str.contains("Action")) & (movies["imdb_rating"] > 7.5)].shape[0]
```
