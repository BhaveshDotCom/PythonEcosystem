```python
expense = pd.read_csv(os.getenv("EXPENSE_FILE_PATH"))
expense["Date"] = pd.to_datetime(expense["Date"])
expense["Date"].dt.month_name()
expense["Date"].dt.is_quarter_end
```

```python
expense["day"] = expense["Date"].dt.day_name()
expense.groupby("day")["INR"].mean().plot(kind="bar")
```

```python
expense["Month"] = expense["Date"].dt.month_name()
expense.groupby("Month")["INR"].sum().plot(kind="bar")
```
