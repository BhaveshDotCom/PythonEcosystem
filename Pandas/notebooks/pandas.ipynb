#### Series

##### series from list

```python

# string
countries = ["India", "Nepal", "Bhutan", "Cuba", "Russia"]
print(pd.Series(countries))
```

```python
#integer
runs = ([129,99,59,34,100])
print(pd.Series(runs))
```

```python
subjects = ["Programming in C++", "Data Structures and Algorithms", "Programming in Java", "Operating System"]
marks = [97,100,99,98]

grades = pd.Series(marks,index=subjects,name="Bhavesh Score Card")

grades
```

##### Series form Dictionary    

```python
players = {
    "Rohit Sharma": 45, 
    "Virat Kohli": 18,
    "MS Dhoni": 7,
    "Jaspreet Bhumrah": 93 
}

pd.Series(players)
```

***Pandas Series – Class and Object***

1. Series is a class provided by the Pandas library.
2. pd.Series() is the constructor of the Series class.
3. Calling pd.Series() creates a new object (instance) of the Series class.
4. The created object is stored in a variable.

##### Series Attributes

```python
players = {
    "Rohit Sharma": 45, 
    "Virat Kohli": 18,
    "MS Dhoni": 7,
    "Jaspreet Bhumrah": 93 
}
players_data = pd.Series(players, name="Indian Team")
print(players_data,'\n')

print("Attributes")
# size
print("size ",players_data.size)

#dtype
print("dtype ", players_data.dtype)

#name
print("name ", players_data.name)

#isunique
print("is_unique ", players_data.is_unique)
print("pd.Series([1,1,2,3]).is_unique ", pd.Series([1,1,2,3]).is_unique)

#index
print("index ", players_data.index)

#values
print("values ", players_data.values)
print("type(values) ", type(players_data.values))
```
