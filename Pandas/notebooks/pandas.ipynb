***Pandas Series – Class and Object***

1. Series is a class provided by the Pandas library.
2. pd.Series() is the constructor of the Series class.
3. Calling pd.Series() creates a new object (instance) of the Series class.
4. The created object is stored in a variable.

##### Series Attributes

```python
players = {
    "Rohit Sharma": 45, 
    "Virat Kohli": 18,
    "MS Dhoni": 7,
    "Jaspreet Bhumrah": 93 
}
players_data = pd.Series(players, name="Indian Team")
print(players_data,'\n')

print("Attributes")
# size
print("size ",players_data.size)

#dtype
print("dtype ", players_data.dtype)

#name
print("name ", players_data.name)

#isunique
print("is_unique ", players_data.is_unique)
print("pd.Series([1,1,2,3]).is_unique ", pd.Series([1,1,2,3]).is_unique)

#index
print("index ", players_data.index)

#values
print("values ", players_data.values)
print("type(values) ", type(players_data.values))
```

##### Series Methods

```python
# head
bollywood_data.head() # return first 5 rows
bollywood_data.head(7) # return first 7 rows
```

```python
# tail
bollywood_data.tail() # return bottom 5 rows
bollywood_data.tail(3) # return bottom 5 rows
```

```python
# sample
bollywood_data.sample() # return random 1 rows
```

```python
# value_count
bollywood_data.value_counts() # frequency in decending order
```

```python
# sort
kohli_runs_data.sort_values() # acend
kohli_runs_data.sort_values(ascending=False) # decend
```

```python
# method chaining
print(kohli_runs_data.sort_values(ascending = False).head(1).values[0])
#* no change in original series

#! Caution
# print(kohli_runs_data.sort_values(inplace=True, ascending=False))
#* changed original series
kohli_runs_data
```

```python
bollywood_data.sort_values(ascending=False)
bollywood_data.sort_index(ascending=False)
```

##### Series Math Methods

```python
# count
#? Does not count missing values
print(kohli_runs_data.count())

# sum / prod
print(kohli_runs_data.sum())
print(kohli_runs_data.product())

# mean/median/mod/std/var
print("median ",kohli_runs_data.median())
print("mode ",bollywood_data.mode())
print("mean ", kohli_runs_data.mean())
print("std ", kohli_runs_data.std())
print("Var ",kohli_runs_data.var())
```

```python
# min/max
print(kohli_runs_data.min())
print(kohli_runs_data.max())
```

```python
print(kohli_runs_data.describe())
```
