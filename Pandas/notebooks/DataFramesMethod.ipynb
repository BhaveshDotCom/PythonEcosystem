```python
marks_dict = {
    "Point A": [(95, 88), (91, 85), (99, 67)],
    "Point B": [(89, 92), (81, 87), (90, 72)]
}

marks = pd.DataFrame(marks_dict)

def distance(row):
    pt_a = row["Point A"]
    pt_b = row["Point B"]
    return np.sqrt((pt_a[0] - pt_b[0])**2 + (pt_a[1] - pt_b[1])**2)

marks["Marks Difference"] = marks.apply(distance ,axis=1)
marks
```
