```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
tips["size"]*=10
tips.sample(2)
```

```python
tips.plot(kind="scatter",x="total_bill",y="tip",figsize=(8,3),title="total bill vs tip",s="size",c="sex",cmap="viridis")
```

```python
np.random.seed(42)

ROWS = 365

stocks = pd.DataFrame({
    "Date": pd.date_range("2024-01-01", periods=ROWS, freq="D")
})

stocks["TechCorp"] = 100 + np.cumsum(np.random.normal(0.3, 1.5, ROWS))
stocks["FinanceInc"] = 80 + np.cumsum(np.random.normal(0.25, 1.2, ROWS))

stocks.loc[180:, "TechCorp"] -= 35
stocks.loc[240:, "FinanceInc"] += 30

stocks[["TechCorp", "FinanceInc"]] = stocks[
    ["TechCorp", "FinanceInc"]
].round(2)
```

```python
stocks.plot(x="Date", y=["TechCorp", "FinanceInc"])
```
