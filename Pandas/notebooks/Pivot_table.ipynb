```python
import numpy as np
import pandas as pd
import seaborn as sns
import os
from dotenv import load_dotenv
load_dotenv(override=True)
```

A pivot table is a data summarization tool in Pandas that groups data by one or more columns and applies an aggregate function (such as sum, mean, count, max, or min) to the values.

OR

A Pivot Table is a data summarization tool that takes column-wise data as input and groups it into a two-dimensional table, providing a multidimensional summary of the data.

```python
expense_data = pd.read_csv(os.getenv("EXPENSE_FILE_PATH"))
```

```python
# toy data
df = sns.load_dataset('tips')
df.head()
```

```python
# Average bill payed on the basis of gender
df.groupby(["sex"])["total_bill"].mean().reset_index()
# # Average bill payed on the basis of gender and smoker
df.groupby(["sex","smoker"])["total_bill"].mean().reset_index()
df.groupby(["sex","smoker"])["total_bill"].mean().unstack()
```

```python
df.pivot_table(index="sex",columns="smoker" ,values="total_bill",aggfunc="sum") # default aggfunc = 'mean'
```

```python
# multidimensional
df.pivot_table(index=["sex", "smoker"],columns=["day","time"], aggfunc=({
    'tip':"max",
    "total_bill": "sum"
}),values=["tip", "total_bill"]).unstack()
```

```python
df.pivot_table(index="sex",columns="smoker",values="total_bill",aggfunc="sum",margins=True)
```

```python
expense_data["Date"] = pd.to_datetime(expense_data["Date"])
expense_data["Month"] = expense_data["Date"].dt.month_name()
expense_data.pivot_table(index="Month",columns="Category",values="INR", aggfunc="sum",fill_value=0).plot()
```

```python
expense_data.pivot_table(index="Month",columns="Income/Expense",values="INR", aggfunc="sum",fill_value=0).plot()
```
