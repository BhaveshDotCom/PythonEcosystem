```python
expense_data = pd.read_csv(os.getenv("EXPENSE_FILE_PATH"))
titanic_data = pd.read_csv(os.getenv("TITANIC_FILE_PATH"))
```

```python
# all name to upper/lower/capitalze case
titanic_data["Name"].str.upper()
```

```python
# longest name
titanic_data[titanic_data["Name"].str.len() == titanic_data["Name"].str.len().max()]["Name"].values[0]
```

```python
titanic_data['last_name'] = titanic_data["Name"].str.split(',').str.get(0)
titanic_data['first_name'] = titanic_data["Name"].str.split(',').str.get(1).str.strip().str.split(" ",n=1).str.get(1)
titanic_data['honorifics'] = titanic_data["Name"].str.split(',').str.get(1).str.strip().str.split(" ",n=1).str.get(0)
titanic_data['honorifics'] = titanic_data['honorifics'].str.replace("Miss.","Ms.")
titanic_data['honorifics'] = titanic_data['honorifics'].str.replace("Mile.","Ms.")
titanic_data['honorifics'] = titanic_data['honorifics'].str.replace("Master","Mr.")
titanic_data['honorifics'] = titanic_data['honorifics'].str.replace("Mr..","Mr.")

titanic_data["honorifics"].value_counts()
```

```python
titanic_data[titanic_data["first_name"].str.contains("john",case=False)]["first_name"]
```

```python
# last_name starts and ends with vowel 
titanic_data[titanic_data["last_name"].str.contains('^[aeiouAEIOU].+[aeiouAEIOU]$')]["first_name"]
# last_name does not starts and ends with vowel 
titanic_data[titanic_data["last_name"].str.contains('^[^aeiouAEIOU].+[^aeiouAEIOU]$')]["first_name"]
```
