##### Broadcasting

Broadcasting in NumPy is a mechanism that allows arrays with different shapes to participate in arithmetic operations without explicitly copying data.

Instead of requiring arrays to have exactly the same shape, NumPy "stretches" the smaller array conceptually to match the larger one.

***Rule 1: Dimensions are Equal***

- If the sizes of the two dimensions are the same, they are compatible.

***Rule 2: One of the Dimensions is 1***

- If one dimension is 1, NumPy stretches (broadcasts) that dimension to match the other array.

***Rule 3: Missing Dimensions are Treated as 1***

- If one array has fewer dimensions, NumPy adds **leading/head** dimensions of size 1 until both arrays have the same number of dimensions.

***Rule 4: If None of the Above Rules Apply, Broadcasting Fails***

- If the dimensions are different and neither is 1, NumPy raises an error.



```python
# mathematically we can not add 2 arrays of different order
arr_one = np.arange(6).reshape(2,3)
print("Array One\n", arr_one, end="\n\n")
arr_two = np.arange(3).reshape(1,3)
print("Array One\n", arr_two, end="\n\n")

print(arr_one+arr_two) # Reason: Broadcasting
```

```python
arr_one = np.random.randint(1,100,6).reshape(2,3)
print("Array One\n", arr_one, end="\n\n")
arr_two = np.random.randint(1,100,1).reshape(1)
print("Array One\n", arr_two, end="\n\n")

'''
arr_one.shape() = (2,3) 
arr_two.shape() = (1) -> (1,1) -> (2,3)
'''
print("Broadcasted Array\n", arr_one+arr_two)
```
